# for loop to export files with different PGS weights

## afr eGFR

```python
df = afr_egfr_snp.copy()
pgs_base_name = 'AFR.eGFR.PRScs.'
trait = 'eGFR'

for col in list(range(5, 105)):
    sub = df[['chr_name', 'chr_position', 'effect_allele', 'other_allele', col]]
    sub = sub.rename(columns = {col : 'effect_weight'})

    index = col - 4
    comments = ['#pgs_name=' + pgs_base_name + str(index),
                '#pgs_id=' + pgs_base_name + str(index),
                '#trait_reported=' + trait,
                '#genome_build=GRCh38']
    output_file = 'PGS_variability/CKD/PRScs/reformatted_outputs/' + pgs_base_name + str(index) + '.PGSC_CALC.txt'
    with open(output_file, 'w') as f:
        for line in comments:
            f.write(line + '\n')
        sub.to_csv(f, sep = '\t', index = False)
```
