## create median absolute deviance function

```python
def mad_filter(df, threshold = 5):
    values = df['value_as_number'].dropna()
    
    # Compute median and MAD
    median = values.median()
    mad = np.median(np.abs(values - median))
    
    # Define limits
    lower_limit = median - threshold * mad
    upper_limit = median + threshold * mad

    # Filter and add top/bottom limits
    filtered_df = df[(df['value_as_number'] >= lower_limit) & (df['value_as_number'] <= upper_limit)].copy()

    return filtered_df
```

## calculate eGFR

### creatinine based

equation
- eGFRcr = 142 x min(Scr/, 1)^ x max(Scr/, 1)^-1.200 x 0.9938^Age x 1.012 [if female]
- Scr = standardized serum creatinine in mg/dL
-  = 0.7 (females) or 0.9 (males)
-  = -0.241 (female) or -0.302 (male)
- min(Scr/, 1) is the minimum of Scr/ or 1.0
- max(Scr/, 1) is the maximum of Scr/ or 1.0
- Age (years)

#### merge w demo df

```python
cre_serum_demo = cre_serum_mean.merge(person_sex, on = 'person_id', how = 'inner')
print(len(cre_serum_mean.index))
print(len(cre_serum_demo.index))
cre_serum_demo.head()
```

```python
cre_serum_demo['AGE'] = cre_serum_demo['measurement_datetime'] - cre_serum_demo['date_of_birth']
cre_serum_demo['AGE'] = cre_serum_demo['AGE'].astype(str).str.replace(r' .*', '', regex = True)
cre_serum_demo['AGE'] = cre_serum_demo['AGE'].astype(float)
cre_serum_demo['AGE'] = cre_serum_demo['AGE']/365.25
cre_serum_demo.head()
```

```python
cre_serum_pos = cre_serum_demo[cre_serum_demo['sCRE'] > 0]
print(len(cre_serum_demo.index))
print(len(cre_serum_pos.index))
```

### remove outlier values

#### create median absolute deviance function

```python
def mad_filter(df, threshold = 5):
    values = df['value_as_number'].dropna()
    
    # Compute median and MAD
    median = values.median()
    mad = np.median(np.abs(values - median))
    
    # Define limits
    lower_limit = median - threshold * mad
    upper_limit = median + threshold * mad

    # Filter and add top/bottom limits
    filtered_df = df[(df['value_as_number'] >= lower_limit) & (df['value_as_number'] <= upper_limit)].copy()

    return filtered_df
```
